```python
from tkinter import *
from tkinter import ttk, filedialog
import tkinter as tk
from tkinter import ttk
import pandas as pd
import numpy as np
import os

###########################################################################

def show_df(df) :
    treeview['column'] = ('1', '2', '3', '4')
    treeview.column("#0", width = 120 )
    treeview.column(1, anchor = "w", width = 120 )
    treeview.column(2, anchor = "w", width = 120 )
    treeview.column(3, anchor = "w", width = 120 )
    treeview.column(4, anchor = "w", width = 120 )
    
    treeview.heading("#0", text = 'album/track', anchor = "center")
    treeview.heading('1', text = 'artist', anchor = "center")
    treeview.heading('2', text = 'track', anchor = "center")
    treeview.heading('3', text = 'member', anchor = "center")
    treeview.heading('4', text = 'etc', anchor = "center")

    df_rows = df.to_numpy().tolist()

    for row_now in df_rows :
        row = tuple(row_now)
        
        if row[0] == 'p' :
            treeview.insert(parent = "", index = 'end', iid = row[2], text = row[3], values = row[4 :])
            
        else :
            treeview.insert(parent = row[0], index = 'end', iid = row[2], text = row[3], values = row[4 :])
            
        if row[0] == 'p' :
            treeview.item(row[2], open = True)
            
    treeview.selection_set(1)
    treeview.see(1)

#     treeview.pack()
    treeview.pack(expand=True, fill="both")
    
def file_open() :
    maindir = 'C:\\Users\\joo09\\Documents\\GitHub\\tkinter_gui\\saved_excel'
    filename = filedialog.askopenfilename(
        initialdir = maindir,
        title = "Open A File",
        filetype = (("xlsx files", "*.xlsx"), ("All Files", "*.*"))
        )
    if filename :
        try :
            filename = r"{}".format(filename)
            df = read_excel(filename)
        except ValueError :
            my_label.config(text = "File couldn't be open")
        except FileNotFoundError :
            my_label.config(text = "File not found")
    
    #Clear old treeview
    clear_tree()
    
    #Set up new treeview
    
    treeview['column'] = ('1', '2', '3', '4')
    treeview.column("#0", width = 120 )
    treeview.column(1, anchor = "w", width = 120 )
    treeview.column(2, anchor = "w", width = 120 )
    treeview.column(3, anchor = "w", width = 120 )
    treeview.column(4, anchor = "w", width = 120 )
    
    
    treeview.heading("#0", text = 'album/track', anchor = "center")
    treeview.heading('1', text = 'artist', anchor = "center")
    treeview.heading('2', text = 'track', anchor = "center")
    treeview.heading('3', text = 'member', anchor = "center")
    treeview.heading('4', text = 'etc', anchor = "center")

    df_rows = df.to_numpy().tolist()

    count = 2
    for row_now in df_rows :
        row = tuple(row_now)
        if row[0] == 'p' :
            treeview.insert(parent = "", index = 'end', iid = int(row[2]), text = row[3], values = row[4 :])
            
        elif row[0] != 'p' :
            treeview.insert(parent = int(row[0]), index = 'end', iid = int(row[2]), text = row[3], values = row[4 :])
            
        if row[0] == 'p' :
            treeview.item(int(row[2]), open = True)
            
    treeview.selection_set(1)
    treeview.see(1)

#     treeview.pack()
    treeview.pack(expand=True, fill="both")
    
def clear_tree() :
    treeview.delete(*treeview.get_children())
    
    
    
def add() :
    addnum = treeview.selection()
    df_copy = df.copy()
    df.reset_index(drop = True, inplace = True)
    addind = int(addnum[0]) - 1

    add_line = pd.DataFrame(columns = df.columns)

    # elif addnum != 1 :
    if df.loc[addind, 'parent'] == 'p' :
        # child to the 'addind' parent
        add_line.loc[0, 'parent'] = df.loc[addind, 'iid']
        add_line.loc[0, 'index'] = 'end'
        add_line.loc[0, 'iid'] = df.loc[addind, 'iid'] + 1
        add_line.loc[0, 'artist'] = df.loc[addind, 'artist']

    elif df.loc[addind, 'parent'] != 'p' :
        # 'addind' was child, same info with adj child
        add_line.loc[0, 'parent'] = df.loc[addind, 'parent']
        add_line.loc[0, 'index'] = 'end'
        add_line.loc[0, 'iid'] = df.loc[addind, 'iid'] + 1
        add_line.loc[0, 'artist'] = df.loc[addind, 'artist']

    
    if addind != df.shape[0] :
        df_temp = df.loc[addind + 1:, :].copy()
        df_temp.reset_index(drop = True, inplace = True)
        push_ind = df_temp.shape[0]

    for i in range(df_temp.shape[0]) :
        df_temp.loc[i, 'iid'] =  df_temp.loc[i, 'iid'] + 1
        
        if df_temp.loc[i, 'parent'] != 'p' :
            df_temp.loc[i, 'parent'] = df_temp.loc[i, 'parent'] + 1
        
    if addind != df.shape[0] :
        for i in range(df_temp.shape[0]) :
            df.loc[addind + 2 + i, :] = None
            df.loc[addind + 2 + i, :] = df_temp.loc[i, :]
            
    df.loc[addind + 1, :] = add_line.loc[0, :]

    for i in range(df.shape[0]) :
        if df.loc[i, 'parent'] != 'p' :
            df.loc[i, 'parent'] = int(df.loc[i, 'parent'])
        df.loc[i, 'iid'] = int(df.loc[i, 'iid'])
    df.reset_index(drop = True, inplace = True)
    print(df_temp)
    print(addind)
    print(df)
    clear_tree()
    show_df(df)
    
def button() :
    pass

def read_excel(name) :
    df = pd.read_excel(name)
            
    if 'Unnamed: 0' in df.columns :
        df.drop('Unnamed: 0', axis = 1, inplace = True)

    if 'Unnamed: 0.1' in df.columns :
        df.drop('Unnamed: 0.1', axis = 1, inplace = True)
        
    return df

###########################################################################

maindir = 'C:\\Users\\joo09\\Documents\\GitHub\\tkinter_gui\\saved_excel'
os.chdir(maindir)
for excel in os.listdir(maindir) :
    df = read_excel(excel)

###########################################################################   
    
orgdir = 'C:\\Users\\joo09\\Documents\\GitHub\\tkinter_gui'
os.chdir(orgdir)
root = tk.Tk()
root.title('GUI')
root.option_add('*tearOff', False)


style = ttk.Style(root)
root.tk.call('source', 'forest-dark.tcl')
style.theme_use('forest-dark')

root.title('Music')
# root.geometry("700x500")

my_frame = Frame(root)
my_frame.pack(pady = 20)

# treeview = ttk.Treeview(my_frame)
treeScroll = ttk.Scrollbar(my_frame)
treeScroll.pack(side="right", fill="y")

treeview = ttk.Treeview(my_frame, selectmode="extended", yscrollcommand=treeScroll.set, columns=(1, 2), height=12)

treeScroll.config(command=treeview.yview)
show_df(df)
            
my_menu = Menu(root)
root.config(menu = my_menu)

file_menu = Menu(my_menu, tearoff = False)
my_menu.add_cascade(label = "Spreadsheets", menu = file_menu)
file_menu.add_command(label = "Open", command = file_open)

Button(root, text="Add", command=add).pack()
print(df)
root.mainloop()
```
